## Matplotlib 범례 표시하기
- 링크 : https://codetorial.net/matplotlib/set_legend.html
- 기타 매개변수 : https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html
- 범례 (Legend)는 그래프에 데이터의 종류를 표시하기 위한 텍스트 이다.
- matplotlib.pyplot 모듈의 legend() 함수를 사용해서 그래프에 범례를 표시할 수 있다.
- Keyword : plt.legend(), label, loc, ncol, fontsize, frameon, shadow, 범례

### 1) 기본 사용
- 예제

- 그래프 영역에 범례를 나타내기 위해서 plot() 함수에 label 문자열을 지정한다.
- matplotlib.pyplot 모듈의 legend() 함수를 호출하면 label에서 지정한 문자열이 범례로 출력된다.

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[2,3,5,10], label='Price ($)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axiis')
plt.legend()

plt.show()
```

### 2) 위치 지정하기
- 예제1
    - legend() 함수의 loc 파라미터를 이용해서 범례가 표시될 위치를 설정한다.
    - loc 파라미터를 숫자 쌍 튜플로 지정하면, 해당하는 위치에 범례가 표시된다.
    - loc=(0.0,0.0)은 데이터 영역의 왼쪽 아래, loc=(1.0,1.0)은 데이터 영역의 오른쪽 위 위치이다.

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[2,3,4,10], label='Price ($)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
# plt.legend(loc=(0.0,0.0))
# plt.legend(loc=(0.5,0.5))
plt.legend(loc=(1.02,0.45))

plt.show()
```

- 예제2
    - loc 파라미터는 예제에서와 같으 문자열로 지정할 수도 있고, 숫자 코드를 사용할 수도 있다.
    - loc='lower right'와 같이 지정하면 아래와 같이 오른쪽 아래에 범례가 표시된다.
    - loc 설정값은 다음과 같습니다.

![nn](./image/step4/legend_location.PNG)

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[2,3,5,10], label='Price ($)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc='lower right')

plt.show()
```

### 3) 열 개수 지정하기
- 예제
    - legend() 함수의 ncol 파라미터는 범례에 표시될 텍스트의 열의 개수를 지정합니다.
    - 기본적으로 ncol 는 best 값을 가지며 1개의 열로 표시됩니다.
    - ncol=2 로 지정시 열이 두개 형태로 나옵니다.

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4], [2,3,5,10], label='Price ($)')
plt.plot([1,2,3,4], [3,5,9,7], label='Demand (#)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
# plt.legend(loc='best')
plt.legend(loc='best', ncol=2)

plt.show()
```

### 4) 폰트 크기 지정하기
- 예제
    - legend() 함수의 fontsize 파라미터는 범례에 표시될 폰트의 크기를 지정합니다.
    - 폰트 크기를 14로 지정

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[2,3,5,10], label='Price ($)')
plt.plot([1,2,3,4],[3,5,9,7], label='Demand (#)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc='best')
plt.legend(loc='best', ncol=2, fontsize=14)

plt.show()
```

### 5) 범례 테두리 꾸미기
- 예제
    - frameon 파라미터는 범례 텍스트 상자의 테두리를 표시할지 여부를 지정합니다.
    - frameon=False로 지정하면 테두리가 표시되지 않습니다.
    - shadow 파라미터를 사용해서 텍스트 상자에 그림자를 표시할 수 있습니다.
    - 이외에도 legend() 함수에는 facecolor, edgecolor, borderpad, labelspacing과 같은 다양한 파라미터가 존재

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4], [2,3,5,10], label='Price ($)')
plt.plot([1,2,3,4],[3,5,9,7], label='Demand (#)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
# plt.legend(loc='best')
plt.legend(loc='best', ncol=2, fontsize=10, frameon=False, shadow=True)

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4], [2,3,5,10], label='Price ($)')
plt.plot([1,2,3,4],[3,5,9,7], label='Demand (#)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
# plt.legend(loc='best')
plt.legend(loc='best', ncol=2, fontsize=10, facecolor='r', shadow=True)

plt.show()
```
