## Matplotlib 마커 지정하기
- 링크 : https://codetorial.net/matplotlib/set_marker.html
- 특별한 설정이 없으면 그래프가 실선이다.
- plot() 함수의 포맷 문자열 (Format string)을 사용해서 그래프의 선과 마커를 지정하는 방법에 대해 알아본다.
- Keyword : plt.plot(), 포맷 문자열, 마커, marker

### 1) 기본사용
- 예제
    - plot() 함수에 숫자 데이터와 함께 'bo'를 입력하면 그래프가 파란색의 원형 마커로 표시된다.
    - 'b'는 blue, 'o'는 circle을 나타내는 문자이다.

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[2,3,5,10],'bo')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

### 2) 선/마커 동시에 나타내기
- 예제
    - 'bo-'는  파란색의 원형 마커와 실선을 의미
    - 'bo--'는 파란색의 원형 마커와 점선을 의미

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[2,3,5,10],'b>:')    # 파란색 + 마커 + 실선
#plt.plot([1,2,3,4],[2,3,5,10],'bo--')   # 파란색 + 마커 + 점선
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

### 4) marker 파라미터 사용하기
- 예제

```python
import matplotlib.pyplot as plt

plt.plot([4,5,6], marker="H")
plt.plot([3,4,5], marker="d")
plt.plot([2,3,4], marker="x")
plt.plot([1,2,3], marker=11)
plt.plot([0,1,2], marker="$Z$")
plt.show()
```
